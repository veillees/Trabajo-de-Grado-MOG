```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Se asumen unidades naturales donde G_N = 1 y G = (1 + alpha)
#alpha acotado entre 

M = 1
alpha = 0.0 # alpha = 0 : General Relativity, alpha_crit = 0.673
G_N = 1; G = G_N*(1+alpha)
delta = 1
```

```python
def Veff_MOG(r, l, delta):
    gamma_MOG = (1 - 2*G*M/r + (alpha*G_N*G**M**2)/r**2)
    return gamma_MOG*(l**2/r**2 + delta) - delta

def Veff_MOG_nu(r, l, delta):
    gamma_MOG_nu = (1 - 2*M*(1+alpha)/r + (alpha*(1+alpha)*M**2)/r**2)
    return gamma_MOG_nu*(l**2/r**2 + delta) - delta

def Veff_Sch(r, l, delta):
    gamma_Sch = (1 - 2*G*M/r)
    return gamma_Sch*(l**2/r**2 + delta) - delta
```

```python

rr = np.linspace(2,50,1000)
l0, l1, l2, l3 = 10, 7, 5, 3
```

```python
plt.figure(dpi=600)
plt.plot(rr, Veff_MOG(rr,l0,delta), color= 'cadetblue', label= str(l0))
plt.plot(rr, Veff_MOG(rr,l1,delta), color= 'crimson', label= str(l1))
plt.plot(rr, Veff_MOG(rr,l2,delta), color= 'olive', label= str(l2))
plt.plot(rr, Veff_MOG(rr,l3,delta), color= 'rebeccapurple', label= str(l3))

#plt.plot(rr, Veff_Sch(rr,l), color= 'cornflowerblue', label='Schwarzchild')
#plt.axhline(0, color='black', alpha=0.3)
#plt.axhline(1, color='black', alpha=0.3)
#plt.axvline(3, color='black', alpha=0.7)
#plt.title('General Relativity $V_{eff}$')
plt.xlabel(r'$r$'); plt.ylabel(r'RG $V_{eff} (r)$')
plt.xlim(0,50); plt.ylim(-1,4.99)
#plt.yscale('log')
plt.grid();
plt.legend(ncol = 2); plt.show()
```

```python
# Se asumen unidades naturales donde G_N = 1 y G = (1 + alpha)
#alpha acotado entre 

M = 1
alpha = 0.5 # alpha = 0 : General Relativity, alpha_crit = 0.673
G_N = 1; G = G_N*(1+alpha)
delta = 1
jj = 0.5

def Veff_MOG(r, l, delta):
    gamma_MOG = (1 - 2*G*M/r + (alpha*G_N*G**M**2)/r**2)
    return gamma_MOG*(l**2/r**2 + delta) - delta

def Veff_MOG_nu(r, l, delta):
    gamma_MOG_nu = (1 - 2*M*(1+alpha)/r + (alpha*(1+alpha)*M**2)/r**2)
    return gamma_MOG_nu*(l**2/r**2 + delta) - delta

def Veff_Sch(r, l, delta):
    gamma_Sch = (1 - 2*G*M/r)
    return gamma_Sch*(l**2/r**2 + delta) - delta

import numpy as np

def Veff_KM(r, a, ell, alpha, m, M):
    # Definir Delta en términos de r, alpha, M, y a
    Delta = r**2 - 2 * (alpha + 1) * M * r + a**2 + alpha * (alpha + 1) * M**2

    # Definir las funciones A(r), B(r) y D(r)
    A = (r**2 + a**2)**2 - a**2 * Delta
    B = (ell * a + alpha * m * M * r) * (r**2 + a**2) - ell * a * Delta
    D = (ell * a + alpha * m * M * r)**2 - Delta * ell**2 - m**2 * r**2 * Delta

    # Calcular y retornar V(r)
    V_r = (B + np.sqrt(B**2 - A * D)) / A
    return V_r





plt.figure(dpi=400)
plt.plot(rr, Veff_KM(rr,jj,l0,delta), color= 'cadetblue', label= str(l0))
plt.plot(rr, Veff_KM(rr,jj,l1,delta), color= 'crimson', label= str(l1))
plt.plot(rr, Veff_KM(rr,jj,l2,delta), color= 'olive', label= str(l2))
plt.plot(rr, Veff_KM(rr,jj,l3,delta), color= 'rebeccapurple', label= str(l3))

#plt.plot(rr, Veff_Sch(rr,l), color= 'cornflowerblue', label='Schwarzchild')
#plt.axhline(0, color='black', alpha=0.3)
#plt.axhline(1, color='black', alpha=0.3)
#plt.axvline(3, color='black', alpha=0.7)
#plt.title('MOG $V_{eff}$')
plt.xlabel(r'$r$'); plt.ylabel(r'MOG (0.5) $V_{eff} (r)$')
plt.xlim(0,50); plt.ylim(-1,4.99)
#plt.yscale('log')
plt.grid(); plt.legend(ncol = 2); plt.show()
```
